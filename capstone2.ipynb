```python
import streamlit as st
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
from pandas.api.types import is_numeric_dtype,is_string_dtype,is_object_dtype
import io
```

```python
df = pd.read_csv("C:/Users/Elsayed Hassan/.spyder-py3/autosave/International_Report_Departures.csv")
```

```python
df = pd.DataFrame()

def load_df(df):
    return df

def is_string_type(s: pd.Series):
    return is_string_dtype(s)

def is_numeric(col: pd.Series):
    try:
        return is_numeric_dtype(col)
    except:
        return False
```

```python
def count_numerical_categorical_column(df,type_to_count):
    count_n = 0
    count_c = 0
    for c in df.columns:
        if type_to_count == "numerical":
            if is_numeric(df[c]):
                count_n +=1
                return count_n
            else:
                count_c +=1
            return count_c
```

```python
 ############fill null values########################
def clean_df(df):
    for c in df:
        if is_numeric(df[c]):
            mean_value = df[c].mean()
            df[c].fillna(mean_value,inplace=True)
        elif is_string_type(df[c]):
            col_mode_value = df[c].mode()
            df[c].fillna(col_mode_value,inplace=True)
    return df
```

```python

def get_df_mumerical_cols(df):
    mumerical_cols = []
    for c in df:
        if is_numeric(df[c]):
            mumerical_cols.append(c)
    return mumerical_cols
```

```python
def get_df_other_cols(df):
    other_cols = []
    for c in df:
        if is_string_dtype(df[c]):
            other_cols.append(c)
    return other_cols
```

```python
def get_df_info(df):
    buffer = io.StringIO()
    df.info(buf=buffer)
    s = buffer.getvalue()
    return s
```

```python
def get_df_summary(df):
    return df.describe()

def get_df_columns(df):
    return df.columns

def get_df_col_types(df):
    return df.dtypes
```
